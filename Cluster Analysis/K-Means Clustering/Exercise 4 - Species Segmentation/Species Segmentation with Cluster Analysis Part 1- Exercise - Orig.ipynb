## Plot the data

For this exercise, try to cluster the iris flowers by the shape of their sepal.

# Clustering (unscaled data)

Separate the original data into 2 clusters.

```python
x = raw_data.copy()
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = raw_data.copy()
data_with_clusters['Cluster'] = identified_clusters
data_with_clusters
```

```python
plt.scatter(data_with_clusters['sepal_length'], data_with_clusters['sepal_width'], c=data_with_clusters['Cluster'], cmap='rainbow')
plt.xlabel('Lenght of sepal')
plt.ylabel('Width of sepal')
plt.show()
```

# Standardize the variables

Import and use the <i> method </i> function from sklearn to standardize the data. 

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(raw_data)
x_scaled
```

# Clustering (scaled data)

```python
kmeans_scaled = KMeans(2)
kmeans_scaled.fit(x_scaled)
```

```python
identified_clusters_scaled = kmeans_scaled.fit_predict(x_scaled)
identified_clusters_scaled
```

```python
data_with_clusters_scaled = raw_data.copy()
data_with_clusters_scaled['cluster_pred'] = identified_clusters_scaled
data_with_clusters_scaled
```

```python
plt.scatter(data_with_clusters_scaled['sepal_length'], data_with_clusters_scaled['sepal_width'], c=data_with_clusters_scaled['cluster_pred'], cmap='rainbow')
plt.xlabel('Lenght of sepal')
plt.ylabel('Width of sepal')
plt.show()
```

## Take Advantage of the Elbow Method

### WCSS

```python
wcss = []

for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
    
wcss
```

### The Elbow Method

```python
number_of_clusters = range(1,10)
plt.plot(number_of_clusters, wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

How many clusters are there?

It seems like 3 or 5-cluster solutions are the best.
