```python
# Transform Yes and No into 0 and 1
data = raw_data.copy()
data['Admitted'] = data['Admitted'].map({'Yes': 1, 'No': 0})
data.head(10)
```

```python
y = data['Admitted']
x1 = data['SAT']
```

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y, x)
results_log = reg_log.fit()
```

```python
# Get the regression summary
results_log.summary()
```
