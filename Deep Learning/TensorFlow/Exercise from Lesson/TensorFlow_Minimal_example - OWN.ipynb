```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
observations = 1000

xs = np.random.uniform(low=-10, high=10, size=(observations,1))
zs = np.random.uniform(low=-10, high=10, size=(observations,1))
generated_inputs = np.column_stack((xs,zs))

noise = np.random.uniform(-1, 1, (observations,1))

generated_targets = 2*xs - 3*zs + 5 + noise

np.savez('E:\\Trainings\\Complete Data Science Bootcamp - Udemy\\Deep Learning\\TensorFlow\\Exercise from Lesson\\TF_intro', inputs=generated_inputs, targets=generated_targets)
```

```python
training_data = np.load("E:\\Trainings\\Complete Data Science Bootcamp - Udemy\\Deep Learning\\TensorFlow\\Exercise from Lesson\\TF_intro.npz")
```

```python
input_size = 2 # as we have 2 input variables xs and zs
output_size = 1

model = tf.keras.Sequential([
                            tf.keras.layers.Dense(output_size,
                                kernel_initializer = tf.random_uniform_initializer(minval=-0.1, maxval=0.1),
                                bias_initializer = tf.random_uniform_initializer(minval=-0.1, maxval=0.1)
                            )
                            ])

custom_optimizer = tf.keras.optimizers.SGD(learning_rate=0.02)
```

```python
model.compile(optimizor=custom_optimizer, loss='mean_squared_error')
```

```python
model.fit(training_data['inputs'], training_data['targets'], epochs=100, verbose=2)
```
