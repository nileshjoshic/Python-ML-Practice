## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

## Load the data

```python
data = pd.read_csv("E:\\Trainings\\Complete Data Science Bootcamp - Udemy\\Regression\\Simple Linear Regression\\Exercise\\real_estate_price_size.csv")
```

```python
data.describe()
```

## Create the regression

### Declare the dependent and the independent variables

```python
y = data['price']
x1 = data['size']
```

### Explore the data

```python
plt.scatter(x1,y)
plt.xlabel("Size",fontsize=20)
plt.ylabel("Price",fontsize=20)
plt.plot()

# we can see below that as the size increases, price increases
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit() # OLS = Ordinary Least Squares
results.summary()
```

### Plot the regression line on the initial scatter

```python
# Create a scatter plot
plt.scatter(x1,y)
# Define the regression equation, so we can plot it later
#yhat = 0.0017*x1 + 0.275
yhat = x1*223.1787+101900
# Plot the regression line against the independent variable (Size)
fig = plt.plot(x1,yhat, lw=4, c='orange', label ='regression line')
# Label the axes
plt.xlabel('Size', fontsize = 20)
plt.ylabel('Price', fontsize = 20)
plt.show()
```
