### Numpy:
Anyone who uses python for scientific computing, uses NumPy <br>
It's a third-party package, allowing us to work with multidimensional arrays (powerful way to organize and process data) <br>
Fundamental library for manipulating larger chunks of information

### PANDAS:
Enhances NumPy further <br>
It allows us to organize data in a tabular form and to attach descriptive labels to the rows and columns, not just numbers as it's with NumPy <br>
In addition, Pandas is endowed with broad set of tools, facilitating our work with various data formats and missing data <br>
Essential package, if we want to to DS with Python <br>
When using Pandas, the data in which we read the csv file, will automatically get converted into a dataframe

### Matplotlib:
A 2D plotting library, specially designed for visualization of NumPy computations <br>
Provides large set of tools, which allows us to adjust Graphs as per our requirements

### SciPy:
Numpy, Pandas and Matplotlib as part of the larger SciPy group of libraries <br>
It's a Python ecosystem containing lots of tools for scientific calculations, suitable for the fields of mathematics, ML, Engineeting & more

### Seaborn:
Seaborn is a Python visualization library based on Matplotlib <br>
It provides high level interface for drawing attractive statistical graphics <br>
We will code using Matplotlib, but use the beauty & style of Seaborn delivers

### Statsmodels:
A package built on NumPy and SciPy which integrates with Pandas <br>
We will mainly create regression and statsmodels as they provide very good summaries

### scikit-learn:
sklearn is one of the most widely used ML libraries <br>
We will use statsmodels to create and discuss regressions, but we will also provide, the relevant code in sklearn <br>
E.g.: statsmodels is the real deal, however, sklearn makes it easier to understand the real deal
