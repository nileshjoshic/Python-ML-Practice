```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('E:\\Trainings\\Complete Data Science Bootcamp - Udemy\\sklearn\\Simple Linear Regression\\Latest\\1.01.Simple-linear-regression.csv')
```

```python
x = data['SAT']
y = data['GPA']
```

```python
x.shape
```

```python
y.shape
```

```python
x_matrix = x.values.reshape(-1,1)
x_matrix.shape
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_matrix,y)
```

```python
reg.score(x_matrix,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[1740]])
```
